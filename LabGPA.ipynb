```python
# create scatterplot with linear regression

plt.figure(figsize=(20,5))
sns.lmplot(data = df, x='sat_sum', y='fy_gpa', height=4, aspect=3)
plt.title('Correlation beween GPA & SAT Scores')
plt.xlabel('SAT Summary (Math + Verbal)')
plt.ylabel('Final Year GPA')
plt.show()
```

```python
# regression analysis using statsmodels
import statsmodels.api as sm

X=df['sat_sum']
y=df['fy_gpa']
X = sm.add_constant(X)    # add constant term 

# fit regression model
model = sm.OLS(y, X).fit()

# print summary
model.summary()
```

```python
# plot regression

plt.figure(figsize=(20,5))
sns.regplot(x='sat_sum', y='fy_gpa', data=df, ci=None, line_kws={'color':'red'})
plt.title('Correlation beween GPA & SAT Scores')
plt.xlabel('SAT Summary (Math + Verbal)')
plt.ylabel('Final Year GPA')
plt.show()
```

```python
# print regression

coefficients = model.params
print(f'The regression equation is: fy_gpa = {coefficients[0]:.2f} + {coefficients[1]:.2f}*sat_sum')
```
